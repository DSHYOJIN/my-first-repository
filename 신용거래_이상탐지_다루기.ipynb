```python
cc_df['hour'] = pd.to_datetime(cc_df['trans_date_trans_time']).dt.hour
```

```python
cc_df.head()
```

```python
# ê²°ì œì‹œê°„ì„ morning, afternoon, night, eveningìœ¼ë¡œ ë¶„ë¥˜í•˜ê¸° ìœ„í•´ í•¨ìˆ˜ë¥¼ ì •ì˜
def hour_func(x):
    if (x >= 6) & (x < 12):
        return 'morning'
    elif (x >= 12) & (x < 18):
        return 'afternoon'
    elif (x >= 18) & (x < 23):
        return 'night'
    else:
        return 'evening'
```

```python
cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
```

```python
cc_df['hour_cat'].value_counts()
```

```python
all_cnt = cc_df.groupby('cc_num')['amt'].count().reset_index()
```
